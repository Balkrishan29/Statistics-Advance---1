Question 7: What is the significance of confidence intervals in statistical analysis?

Answer. Confidence intervals (CIs) are essential in statistical analysis because they provide a range of values within which the true population parameter (such as a mean or proportion) is likely to fall, with a specified level of confidence (usually 95% or 99%).

Significance of Confidence Intervals:

1. Estimate Uncertainty:
CIs quantify the uncertainty associated with a sample estimate. Instead of giving just a single point estimate (e.g., sample mean), a CI shows the likely range of values for the true population parameter.

2. Informed Decision-Making:
By showing the degree of uncertainty, CIs help researchers, policymakers, and analysts make more informed decisions based on data.

3. Statistical Significance:
Confidence intervals can be used to determine whether a result is statistically significant. For example, if a 95% CI for the difference between two group means does not include zero, it suggests a statistically significant difference.

4. Comparison Tool:
CIs allow for comparison between groups or variables. Overlapping confidence intervals might suggest no significant difference, while non-overlapping intervals might indicate a real effect.

```python
#Question 9: Write a Python program to generate 1000 random numbers from a normal
#distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
#deviation using NumPy, and draw a histogram to visualize the distribution.

import numpy as np
import matplotlib.pyplot as plt

mean = 50
std_dev = 5
data = np.random.normal(loc=mean , scale=std_dev, size=1000)

computed_mean = np.mean(data)
computed_std_dev = np.std(data)

#plot a histogram
plt.hist(data, bins=30, color='skyblue',edgecolor='black')
plt.title('Histogram of random numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

print(f"Computed Mean: {computed_mean}")
print(f"Computed Standard Deviation: {computed_std_dev}")
```

```python
#question 10 pyhton code


import numpy as np
import scipy.stats as stats


daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

data = np.array(daily_sales)

n = len(data)
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)

# Z-score for 95% confidence
z_score = 1.96

margin_of_error = z_score * (sample_std / np.sqrt(n))

# Confidence Interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

```
